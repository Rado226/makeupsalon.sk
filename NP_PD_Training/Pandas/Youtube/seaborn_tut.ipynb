```python
import numpy as np
import pandas as pd
import seaborn as sns 

sns.get_dataset_names()
```

Tutorial na pou≈æitie seaborn 

```python
tips = sns.load_dataset("tips")
tips
titanic = sns.load_dataset("titanic")
```

1. Vytvorenie scatterplot

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="sex", size="size") #palette
```

2. Histogram

```python
sns.histplot(tips["tip"], kde=True, bins=15)
```

3. Bar plot

```python
sns.barplot(x ="sex", y ="tip", data = tips, palette="bright")
```

4. Box plot

```python
sns.boxplot(x="day", y="tip",data=tips, hue="sex")
```

5. Stripplot

```python
sns.stripplot(x="day", y="tip",data = tips, hue="sex", palette="bright", dodge=True)
```

6. Join plot

```python
sns.jointplot(x="tip", y = "total_bill", data=tips, kind="reg")
```

```python
sns.jointplot(x="tip", y = "total_bill", data=tips, kind="hex")
```

```python
sns.jointplot(x="tip", y = "total_bill", data=tips, kind="kde",shade =True)
```

7. Pair Plot

```python
titanic.head(3)
```

```python
sns.pairplot(titanic.select_dtypes(["number"]),hue = "pclass")
```

8. Heatmap

```python
num_titanic = titanic.select_dtypes(include=["number"])
sns.heatmap(num_titanic, annot=True, cmap="YlGnBu")
```
